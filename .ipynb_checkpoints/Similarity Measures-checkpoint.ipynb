```python
import datetime
import pandas as pd
import numpy as np
from sklearn.metrics import jaccard_score
import operator
```

```python
dataTr = pd.read_csv('dataTr.csv')
```

```python
dataTr.columns =['Index', 'Rank', 'Name', 'City', 'State', 'PubOrPriv',
                 'UndergradPop', 'StudentPop', 'NetPrice', 'AverageGrantAid', 'TotalAnnualCost', 'AlumniSalary', 'AcceptanceRate', 'SATLower', 'SATUpper', 'ACTLower', 'ACTUpper', 'Website']
dataTr.head()
```

```python
# Adding extra columns

# Location
Northeast = ['PA', 'NY', 'NJ', 'ME', 'NH', 'VT', 'MA', 'CT', 'RI']
Midwest = ['ND', 'SD', 'NE', 'KS', 'MN', 'IA', 'MO', 'WI', 'IL', 'IN', 'MI', 'OH']
South = ['OK', 'TX', 'AR', 'LA', 'KY', 'TN', 'MS', 'AL', 'WV', 'MD', 'VA', 'DC', 'NC', 'SC', 'GA', 'AL', 'FL', 'DE']
West = ['WA', 'OR', 'CA', 'NV', 'ID', 'MT', 'WY', 'UT', 'AZ', 'NM', 'CO']

# Cost
# $5,000 - $15,000 → 0
# $15,001 - $30,000 → 1
# $30,001 - $45,000 → 2
# $45,001 - $60,000 → 3
# $60,001 - $75,000 → 4

for index, row in dataTr.iterrows():
    size = dataTr.loc[index]['StudentPop']
    if (size < 5000):
        dataTr.loc[index,'Size'] = 'Small'
        dataTr.loc[index,'SizeNum'] = 0
    elif (size >= 5000 and size < 15000):
        dataTr.loc[index,'Size'] = 'Medium'
        dataTr.loc[index,'SizeNum'] = 1
    else:
        dataTr.loc[index,'Size'] = 'Large'
        dataTr.loc[index,'SizeNum'] = 2
    
    state = dataTr.loc[index]['State']
    if (state in Northeast):
        dataTr.loc[index,'Location'] = 'Northeast'
        dataTr.loc[index,'LocationNum'] = 0
    elif (state in Midwest):
        dataTr.loc[index,'Location'] = 'Midwest'
        dataTr.loc[index,'LocationNum'] = 1
    elif (state in South):
        dataTr.loc[index,'Location'] = 'South'
        dataTr.loc[index,'LocationNum'] = 2
    else:
        dataTr.loc[index,'Location'] = 'West'
        dataTr.loc[index,'LocationNum'] = 3
        
    pubPrivStatus = dataTr.loc[index]['PubOrPriv']
    if (pubPrivStatus == 'Private'):
        dataTr.loc[index,'pubOrPrivNum'] = 1
    else:
        dataTr.loc[index,'pubOrPrivNum'] = 0
        
    
    cost = dataTr.loc[index]['TotalAnnualCost']
    if (cost <= 15000):
        dataTr.loc[index,'Cost'] = 0
    elif (cost > 15000 and cost <= 30000):
        dataTr.loc[index,'Cost'] = 1
    elif (cost > 30000 and cost <= 45000):
        dataTr.loc[index,'Cost'] = 2
    elif (cost > 45000 and cost <= 60000):
        dataTr.loc[index,'Cost'] = 3
    else:
         dataTr.loc[index,'Cost'] = 4
```
